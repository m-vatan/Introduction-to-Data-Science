```python
import requests
import urllib
from bs4 import BeautifulSoup as bs
import csv

URL = 'https://www.skinnytaste.com/page/'



food_type = []
food_category = []
img_number = 1
blue = 0
green = 0
purple = 0
calories = 0

csv_file = open ('Food_names.csv', 'w')
csv_writer = csv.writer(csv_file)
csv_writer.writerow(['Food Name', 'Food Calorie', 'Summary', 'Blue Point Value', 'Green Point Value', 'Purple Point Value', 'Key Words', 'img_link'])


for page in range(1,31):
    req = requests.get(URL+str(page))
    soup = bs(req.text, 'html.parser')
    source = soup.find_all('article', class_= ('post teaser-post odd','post teaser-post even'))
    for src in source:
        #Food Names
        food_name = src.h2.text
        #Food Summary
        summary = src.p.text
        
        # Value of Calorie, Blue, Green, and Purple Points
        if src.find('span', class_= 'smart-points blue') is not None:
            blue = src.find('span', class_= 'smart-points blue').text
        else:
            next
           
        if src.find('span', class_= 'smart-points purple') is not None:
            purple = src.find('span', class_= 'smart-points purple').text
        else:
            next
        
        if src.find('span', class_= 'smart-points green') is not None:
            green = src.find('span', class_= 'smart-points green').text
        else:        
            next
        
        if src.find('span', class_= 'icon-star') is not None:
            calories = src.find('span', class_= 'icon-star').text
        else:
            next
    
        #The Keys
        spans = src.find('div', class_=('post-meta', 'icons'))
        s = spans.select('span > a > noscript > img ')
        for s in s:
            food = s.get('alt')
            food_type.append([food])
               
        
        #The image of the food
        img = src.find('img')
        img_lin = img.get('src')
        img_link = requests.get(img_lin)
        
        #Write data on .csv file
        try:
            with open('Image'+str(img_number)+'.jpg', 'wb') as file:
                file.write(img_link.content)
            img_number+=1
            csv_writer.writerow((food_name, calories, summary, blue, green, purple, food_type, img_lin))
        except UnicodeEncodeError:
            pass
        
csv_file.close()
    
 
```
